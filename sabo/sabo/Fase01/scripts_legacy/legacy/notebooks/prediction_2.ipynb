```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_squared_error,
    r2_score,
    mean_absolute_error,
)  # Import mean_absolute_error
```

```python
models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "XGBoost": xgb.XGBRegressor(
        random_state=42, objective="reg:squarederror"
    ),  # Important for regression
}

results = {}

for name, model in models.items():
    print(f"Training and evaluating {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    results[name] = {
        "model": model,
        "mse": mse,
        "r2": r2,
        "mae": mae,  # Added MAE
        "y_pred": y_pred,
        "y_true": y_test,  # Added y_true - CRUCIAL for other metrics!
    }


# Compare the models
print("\nModel Comparison:")
for name, metrics in results.items():
    print(f"{name}:")
    print(f"  MSE: {metrics['mse']:.4f}")
    print(f"  R-squared: {metrics['r2']:.4f}")
    print(f"  MAE: {metrics['mae']:.4f}")  # Print MAE

# Plotting:
plt.figure(figsize=(12, 8))
for i, (name, metrics) in enumerate(results.items()):
    plt.subplot(2, 2, i + 1)
    plt.scatter(X_test[:, 0], y_test, label="Actual", alpha=0.5)  # Use array indexing
    plt.scatter(
        X_test[:, 0], metrics["y_pred"], label=name + " Prediction", alpha=0.5
    )  # Use array indexing
    plt.xlabel("Feature 0")  # Or a more descriptive name if you know it
    plt.ylabel("Target Variable")
    plt.title(name + " Predictions")
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform  # For integer hyperparameters

models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "XGBoost": xgb.XGBRegressor(random_state=42, objective="reg:squarederror"),
}


param_distributions = {  # Distributions for RandomizedSearchCV
    "Random Forest": {
        "n_estimators": randint(50, 1000),  # Random integers between 50 and 200
        "max_depth": [None, 10, 20, 30, 50, None],  # Include None
        "min_samples_split": randint(2, 8),
        "min_samples_leaf": randint(1, 10),
        "max_features": ["sqrt", "log2"],
        "bootstrap": [True, False],
    },
    "Decision Tree": {
        "max_depth": [None, 10, 20, 30, 50, None],
        "min_samples_split": randint(2, 8),
        "min_samples_leaf": randint(1, 10),
        "max_features": ["sqrt", "log2"],
        "splitter": ["best", "random"],
    },
    "XGBoost": {
        "n_estimators": randint(50, 500),
        "max_depth": randint(3, 20),
        "learning_rate": [0.001, 0.01, 0.05, 0.1, 0.3, 0.5],
    },
}

# Store the *search objects* (not just the best estimators):
best_models = {}
model_searches = {}  # A dictionary to store the RandomizedSearchCV objects
scoring = ["neg_mean_squared_error", "r2", "neg_mean_absolute_error"]

for name, _ in models.items():  # Iterate over the keys of 'models'
    print(f"Tuning hyperparameters for {name}...")

    # Update the model within the loop
    if name == "Random Forest":
        model = RandomForestRegressor(random_state=42)
    elif name == "Decision Tree":
        model = DecisionTreeRegressor(random_state=42)
    elif name == "XGBoost":
        model = xgb.XGBRegressor(random_state=42, objective="reg:squarederror")
    print(f"Tuning hyperparameters for {name}...")
    random_search = RandomizedSearchCV(
        model,
        param_distributions[name],
        n_iter=20,  # Number of random combinations to try
        cv=10,
        scoring=scoring,  # Use appropriate scoring
        refit="neg_mean_squared_error",  # Specify which metric to use for refitting the best model
        n_jobs=-1,
        random_state=42,  # For reproducibility
    )
    random_search.fit(X_train, y_train)
    model_searches[name] = random_search  # Store the *search object*
    best_models[name] = random_search.best_estimator_

    print(f"Best hyperparameters for {name}: {random_search.best_params_}")
    print(f"Best score for {name}: {random_search.best_score_}")
```

```python
import seaborn as sns
import plotly.express as px

for name, model_search in model_searches.items():
    print(f"Results for {name}:")
    results_df = pd.DataFrame(model_search.cv_results_)
    print(results_df.head())

    # Find the best row (highest score)
    best_row = results_df.loc[results_df["mean_test_r2"].idxmax()]
    print(f"Best row:\n{best_row}")

    # Scatter plot (n_estimators vs. R-squared)
    if "param_n_estimators" in results_df.columns:
        plt.figure()
        plt.scatter(results_df["param_n_estimators"], results_df["mean_test_r2"])
        plt.xlabel("n_estimators")
        plt.ylabel("Mean R-squared")
        plt.title(f"n_estimators vs. Mean R-squared for {name}")
        plt.show()

        # max_depth vs. R-squared
    if "param_max_depth" in results_df.columns:
        plt.figure()
        plt.scatter(results_df["param_max_depth"], results_df["mean_test_r2"])
        plt.xlabel("max_depth")
        plt.ylabel("Mean R-squared")
        plt.title(f"max_depth vs. Mean R-squared for {name}")
        plt.show()

    # min_samples_split vs. R-squared (if applicable)
    if "param_min_samples_split" in results_df.columns:
        plt.figure()
        plt.scatter(results_df["param_min_samples_split"], results_df["mean_test_r2"])
        plt.xlabel("min_samples_split")
        plt.ylabel("Mean R-squared")
        plt.title(f"min_samples_split vs. Mean R-squared for {name}")
        plt.show()

    # learning_rate vs. R-squared (if applicable)
    if "param_learning_rate" in results_df.columns:
        plt.figure()
        plt.scatter(results_df["param_learning_rate"], results_df["mean_test_r2"])
        plt.xlabel("learning_rate")
        plt.ylabel("Mean R-squared")
        plt.title(f"learning_rate vs. Mean R-squared for {name}")
        plt.xscale("log")  # Use logarithmic scale for learning rate
        plt.show()

    # Parallel Coordinates Plot (using R-squared for color)
    fig = px.parallel_coordinates(
        results_df,
        color="mean_test_r2",
        title=f"Parallel Coordinates Plot for {name}",
    )
    fig.show()

    # Heatmap (using R-squared for values)
    if (
        "param_n_estimators" in results_df.columns
        and "param_max_depth" in results_df.columns
    ):
        pivoted_results = results_df.pivot_table(
            values="mean_test_r2",
            index="param_n_estimators",
            columns="param_max_depth",
        )
        plt.figure()
        sns.heatmap(pivoted_results, annot=True, cmap="viridis")
        plt.title(f"Heatmap of n_estimators vs. max_depth for {name}")
        plt.xlabel("max_depth")
        plt.ylabel("n_estimators")
        plt.show()

    if name == "Random Forest":
        # Heatmap for n_estimators vs. max_depth (Random Forest)
        if (
            "param_n_estimators" in results_df.columns
            and "param_max_depth" in results_df.columns
        ):
            pivoted_results = results_df.pivot_table(
                values="mean_test_r2",
                index="param_n_estimators",
                columns="param_max_depth",
            )
            plt.figure()
            sns.heatmap(pivoted_results, annot=True, cmap="viridis")
            plt.title(f"Heatmap of n_estimators vs. max_depth for {name}")
            plt.xlabel("max_depth")
            plt.ylabel("n_estimators")
            plt.show()

    elif name == "Decision Tree":
        # Heatmap for max_depth vs. min_samples_split (Decision Tree)
        if (
            "param_max_depth" in results_df.columns
            and "param_min_samples_split" in results_df.columns
        ):
            pivoted_results = results_df.pivot_table(
                values="mean_test_r2",
                index="param_max_depth",
                columns="param_min_samples_split",
            )
            plt.figure()
            sns.heatmap(pivoted_results, annot=True, cmap="viridis")
            plt.title(f"Heatmap of max_depth vs. min_samples_split for {name}")
            plt.xlabel("min_samples_split")
            plt.ylabel("max_depth")
            plt.show()

    elif name == "XGBoost":
        # Heatmap for learning_rate vs. max_depth (XGBoost)
        if (
            "param_learning_rate" in results_df.columns
            and "param_max_depth" in results_df.columns
        ):
            pivoted_results = results_df.pivot_table(
                values="mean_test_r2",
                index="param_learning_rate",
                columns="param_max_depth",
            )
            plt.figure()
            sns.heatmap(pivoted_results, annot=True, cmap="viridis")
            plt.title(f"Heatmap of learning_rate vs. max_depth for {name}")
            plt.xlabel("max_depth")
            plt.ylabel("learning_rate")
            plt.show()
```
