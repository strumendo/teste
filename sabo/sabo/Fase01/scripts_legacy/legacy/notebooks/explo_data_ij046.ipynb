```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read the CSV file into a DataFrame
df = pd.read_csv("mnt-oficial-IJ-046.csv")

# Display the first 5 rows
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df.info())
```

```python
# Generate and print descriptive statistics
print(df.describe().to_markdown(numalign="left", stralign="left"))

# Check for duplicate entries
print("Qtd. of duplicates:", df.duplicated().sum())

# Analyze distribution of categorical variables
categorical_vars = [
    "Cód. Recurso",
    "Cód. Produto",
    "Cód. Un.",
    "Descrição da massa (Composto)",
]

for var in categorical_vars:
    print(f"\nDistribution of {var}:")
    print(f"Unique values: {df[var].unique()}")
    print(
        f'Value counts:\n{df[var].value_counts().to_markdown(numalign="left", stralign="left")}'
    )

    # Create bar plots for categorical variables
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, data=df)
    plt.title(f"Distribution of {var}")
    plt.xlabel(var)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```

```python
# Correlation analysis
# Calculate and print the correlation matrix for numeric columns only
correlation_matrix = df.select_dtypes(include=["int64", "float64"]).corr()
print(correlation_matrix.to_markdown(numalign="left", stralign="left"))

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Create scatter plots for variables with the highest correlations
# Get the pairs of variables with the highest absolute correlations (excluding self-correlations)
high_corr_pairs = (
    correlation_matrix.stack()
    .abs()
    .sort_values(ascending=False)
    .drop_duplicates()
    .head(5)  # Adjust the number of pairs as needed
)

for (var1, var2), corr in high_corr_pairs.items():
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=var1, y=var2, data=df)
    plt.title(f"Scatter Plot of {var1} vs. {var2} (Correlation: {corr:.2f})")
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()
```
