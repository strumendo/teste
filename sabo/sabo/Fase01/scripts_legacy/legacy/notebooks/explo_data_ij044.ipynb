````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("mnt-oficial-IJ-044.csv")

# Display the first 5 rows
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df.info())
````

````python
# Generate and print descriptive statistics
print(df.describe().to_markdown(numalign="left", stralign="left"))

# Check for duplicate entries
print("Qtd. of duplicates:", df.duplicated().sum())

# Analyze distribution of categorical variables
categorical_vars = [
    "Cód. Recurso",
    "Cód. Produto",
    "Cód. Un.",
    "Descrição da massa (Composto)",
]

for var in categorical_vars:
    print(f"\nDistribution of {var}:")
    print(f"Unique values: {df[var].unique()}")
    print(
        f'Value counts:\n{df[var].value_counts().to_markdown(numalign="left", stralign="left")}'
    )

    # Create bar plots for categorical variables
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, data=df)
    plt.title(f"Distribution of {var}")
    plt.xlabel(var)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.show()
````

````python
# Correlation analysis
# Calculate and print the correlation matrix for numeric columns only
correlation_matrix = df.select_dtypes(include=["int64", "float64"]).corr()
print(correlation_matrix.to_markdown(numalign="left", stralign="left"))

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Create scatter plots for variables with the highest correlations
# Get the pairs of variables with the highest absolute correlations (excluding self-correlations)
high_corr_pairs = (
    correlation_matrix.stack()
    .abs()
    .sort_values(ascending=False)
    .drop_duplicates()
    .head(5)  # Adjust the number of pairs as needed
)

for (var1, var2), corr in high_corr_pairs.items():
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=var1, y=var2, data=df)
    plt.title(f"Scatter Plot of {var1} vs. {var2} (Correlation: {corr:.2f})")
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()
````

## Principais Insights

*   **Entradas Duplicadas:** Não há entradas duplicadas no conjunto de dados. Isso garante a integridade da análise e evita resultados tendenciosos.

*   **Código do Recurso (Cód. Recurso):** `Cód. Recurso` tem apenas um valor único ('IJ-044'), indicando que todos os dados pertencem a um único recurso. Esta variável não fornece nenhuma informação discriminatória para análise posterior neste contexto.

*   **Código da Unidade (Cód. Un.):** `Cód. Un.` também tem apenas um valor único ('PC'). Semelhante a `Cód. Recurso`, esta variável é constante e não contribui para a compreensão das variações dentro dos dados.

*   **Código do Produto (Cód. Produto):** `Cód. Produto` tem 14 valores únicos, sendo 'SA08395' o mais frequente. Isso sugere que 'SA08395' é o produto mais comumente produzido. Uma investigação mais aprofundada das características deste produto pode ser benéfica.

*   **Descrição da Massa (Composto) (Descrição da massa (Composto)):** `Descrição da massa (Composto)` tem 17 valores únicos, sendo 'N-142/67' o mais frequente. Isso indica que 'N-142/67' é o composto mais utilizado. Analisar as propriedades deste composto e sua relação com a qualidade do produto pode ser valioso.

## Análise de Correlação

A análise de correlação revelou as seguintes relações principais:

*   **Quantidade Produzida vs. Quantidade Refugada (Qtd. Produzida vs. Qtd. Refugada):** Há uma forte correlação positiva entre `Qtd. Produzida` e `Qtd. Refugada`. Isso sugere que, à medida que a quantidade produzida aumenta, o número de refugos também tende a aumentar. Este achado justifica uma investigação mais aprofundada para identificar gargalos potenciais ou ineficiências de processo que contribuem para taxas de refugo mais altas em volumes de produção mais elevados.

*   **Consumo de Massa vs. Quantidade Produzida (Consumo de massa no item em (Kg/100pçs) vs. Qtd. Produzida):** Há uma correlação positiva moderada entre `Consumo de massa no item em (Kg/100pçs)` e `Qtd. Produzida`. Isso indica que quantidades de produção mais altas geralmente estão associadas a um maior consumo de massa. Esta é uma relação esperada, mas a força da correlação pode ser analisada mais a fundo para otimizar o uso da massa e reduzir o desperdício.
