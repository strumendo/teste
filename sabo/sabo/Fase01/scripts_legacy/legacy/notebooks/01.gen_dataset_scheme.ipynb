### 2. Algoritmos de Machine Learning

#### Algoritmos para Regressão

Para prever o consumo de massa (`Consumo de massa no item (em Kg)`), você pode explorar diferentes algoritmos de regressão, incluindo:

**Modelos Lineares:**

* **Regressão Linear:** Um modelo simples e interpretável, que assume uma relação linear entre as variáveis preditoras e a variável alvo.
    ```python
    from sklearn.linear_model import LinearRegression
    modelo = LinearRegression()
    ```

* **Regressão Ridge e Lasso:**  Variantes da regressão linear que adicionam regularização para evitar overfitting, especialmente útil quando há muitas variáveis preditoras.
    ```python
    from sklearn.linear_model import Ridge, Lasso
    modelo_ridge = Ridge(alpha=1.0)  # alpha controla a força da regularização
    modelo_lasso = Lasso(alpha=1.0)
    ```

**Modelos Baseados em Árvores:**

* **Árvores de Decisão para Regressão:**  Constroem uma árvore de decisões para prever valores contínuos. Podem capturar relações não lineares nos dados.
    ```python
    from sklearn.tree import DecisionTreeRegressor
    modelo = DecisionTreeRegressor()
    ```

* **Random Forest:** Combina várias árvores de decisão para melhorar a precisão e a robustez do modelo.
    ```python
    from sklearn.ensemble import RandomForestRegressor
    modelo = RandomForestRegressor(n_estimators=100)  # n_estimators define o número de árvores
    ```

* **XGBoost:**  Um algoritmo de boosting que constrói árvores de decisão sequencialmente, dando mais peso aos erros das árvores anteriores. Geralmente apresenta alta performance.
    ```python
    import xgboost as xgb
    modelo = xgb.XGBRegressor()
    ```

**Outros Modelos:**

* **Support Vector Regression (SVR):**  Utiliza funções kernel para mapear os dados para um espaço de dimensões maiores, onde a regressão linear pode ser aplicada.
    ```python
    from sklearn.svm import SVR
    modelo = SVR(kernel='rbf')  # 'rbf' é um tipo de kernel
    ```

* **Redes Neurais:**  Modelos complexos que podem aprender relações não lineares complexas nos dados. Requerem grande quantidade de dados para treinamento.
    ```python
    from sklearn.neural_network import MLPRegressor
    modelo = MLPRegressor(hidden_layer_sizes=(100,))  # Define a arquitetura da rede
    ```


#### Algoritmos para Classificação e Clustering

* **Árvores de Decisão para Classificação:** Para classificar os produtos (`Cód. Produto`).
    ```python
    from sklearn.tree import DecisionTreeClassifier
    modelo = DecisionTreeClassifier()
    ```

* **K-Means:** Para agrupar os equipamentos (`EQUIPAMENTO`) em clusters.
    ```python
    from sklearn.cluster import KMeans
    modelo = KMeans(n_clusters=5)
    ```


**Observações:**

* A escolha do algoritmo dependerá das características dos seus dados e do problema que você está tentando resolver.
* É crucial realizar a **engenharia de features**, **seleção de variáveis** e **otimização de hiperparâmetros** para obter os melhores resultados.
* **Experimente diferentes algoritmos** e compare seus desempenhos para escolher o modelo mais adequado.


Lembre-se que este é um guia, e a escolha ideal dos algoritmos e do layout dependerá da análise exploratória dos dados e dos seus objetivos específicos.

````python
# Crie um DataFrame com as colunas especificadas e 1000 linhas
df = pd.DataFrame(
    {
        "Data de Produção Acumulada": pd.to_datetime(
            np.random.choice(pd.date_range("2023-01-01", "2023-12-31"), size=1000)
        ),
        "Cod. Ordem": np.random.randint(1000000, 9999999, size=1000),
        "Cod Recurso": np.random.choice(["IJ-046", "IJ-117", "IJ-118"], size=1000),
        "Cod Produto": np.random.choice(["SA05780", "SA02961", "SA02004"], size=1000),
        "Qt. Total Acumulada Produzida até a data específica": np.random.randint(
            1, 1000, size=1000
        ),
        "Qt. Acumulada Refugada até a data específica": np.random.randint(
            1, 100, size=1000
        ),
        "Qtd. Acumulada total Retrabalhada até a data específica": np.random.randint(
            1, 50, size=1000
        ),
        "Fator Un.": np.ones(1000),
        "Cód. Un.": np.full(1000, "PC"),
        "Descrição da massa (Composto)": np.random.choice(
            ["N-142/67", "P212/1"], size=1000
        ),
        "Consumo total de Massa Acumulada": np.random.uniform(0.5, 1.5, size=1000),
    }
)
````

````python
# Calcule a coluna `Tempo Restante para Manutenção`
df["Tempo Restante para Manutenção"] = (
    350 - (pd.to_datetime("today") - df["Data de Produção Acumulada"]).dt.days
)

# Salve o DataFrame em um arquivo CSV
df.to_csv("dados_manutencao.csv", index=False)

# Exiba as primeiras 5 linhas do DataFrame
print("Primeiras 5 linhas do DataFrame:")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Exiba as últimas 5 linhas do DataFrame
print("Últimas 5 linhas do DataFrame:")
print(df.tail().to_markdown(index=False, numalign="left", stralign="left"))

# Exiba o número de valores únicos de cada coluna do DataFrame
print("Número de valores únicos de cada coluna do DataFrame:")
print(df.nunique().to_markdown(index=False, numalign="left", stralign="left"))
````
