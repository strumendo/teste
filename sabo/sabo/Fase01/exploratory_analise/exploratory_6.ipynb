```python
# Após analisar os dados do arquivo "IJ-044.csv", notei que a coluna DATA ENTRADA EM SERVIÇO não está presente. Isso significa que não posso calcular a coluna Dias em Serviço e, portanto, não posso gerar o gráfico de dispersão que inclui essa variável. No entanto, ainda posso gerar os outros gráficos para analisar as variáveis disponíveis.
```

```python
import altair as alt

# Renomeie a coluna "Consumo de massa no item em (Kg/100pçs)"
df_ij044 = df_ij044.rename(
    columns={
        "Consumo de massa no item em (Kg/100pçs)": "Consumo de massa no item(em Kg)"
    }
)

# Recalcule a coluna "Consumo de massa total"
df_ij044["Consumo de massa total"] = (
    df_ij044["Qtd. Produzida"] * df_ij044["Consumo de massa no item(em Kg)"]
)

# Histogramas
histogram_consumo_massa = (
    alt.Chart(df_ij044)
    .mark_bar()
    .encode(
        alt.X("Consumo de massa total", bin=True),
        y="count()",
        tooltip=[alt.Tooltip("Consumo de massa total", bin=True), "count()"],
    )
    .properties(title="Histograma - Consumo de massa total")
    .interactive()
)

histogram_qtd_produzida = (
    alt.Chart(df_ij044)
    .mark_bar()
    .encode(
        alt.X("Qtd. Produzida", bin=True),
        y="count()",
        tooltip=[alt.Tooltip("Qtd. Produzida", bin=True), "count()"],
    )
    .properties(title="Histograma - Qtd. Produzida")
    .interactive()
)

histogram_consumo_massa.save("histograma_consumo_massa_total_ij044.html")
histogram_qtd_produzida.save("histograma_qtd_produzida_ij044.html")

# Boxplot
boxplot_consumo_massa = (
    alt.Chart(df_ij044)
    .mark_boxplot()
    .encode(y="Consumo de massa total")
    .properties(title="Boxplot - Consumo de massa total")
)

boxplot_consumo_massa.save("boxplot_consumo_massa_total_ij044.html")
```

## Análise Exploratória dos Dados: Interpretação dos Gráficos

Os histogramas e o boxplot gerados fornecem informações relevantes sobre a distribuição e variabilidade do consumo de massa e da quantidade produzida para o recurso 'IJ-044'.

* **Histograma de Consumo de Massa Total:** Mostra que a maioria dos valores se concentra em faixas mais baixas, com alguns valores mais altos, o que indica que o consumo de massa pode variar bastante dependendo da produção. Essa variação pode ser influenciada por fatores como o tipo de produto, a massa utilizada e as condições de produção.

* **Histograma de Quantidade Produzida:** Mostra uma distribuição mais uniforme, com a maioria dos valores concentrados em torno de um valor médio. Essa uniformidade sugere que a quantidade produzida é mais consistente ao longo do tempo, com menos variações extremas.

* **Boxplot de Consumo de Massa Total:** Confirma a presença de outliers e mostra a mediana, os quartis e a dispersão dos dados, o que ajuda a entender a variabilidade do consumo de massa. Os outliers podem indicar situações atípicas na produção, que demandaram um consumo de massa maior ou menor do que o habitual.

### Limitação na Análise:

Devido à falta da coluna `DATA ENTRADA EM SERVIÇO` no arquivo "IJ-044.csv", não foi possível gerar o gráfico de dispersão e analisar a relação entre o tempo de uso do equipamento e o consumo de massa.
