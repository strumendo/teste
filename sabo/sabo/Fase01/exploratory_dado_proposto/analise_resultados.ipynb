# Análise dos Resultados da Classificação (Árvore de Decisão)

Para analisar o desempenho do modelo de classificação com Árvore de Decisão, você deve prestar atenção às seguintes métricas:

* **Acurácia:** Indica a porcentagem de acertos do modelo. No entanto, sozinha, ela pode não ser suficiente, principalmente se houver classes desbalanceadas.

* **Relatório de Classificação:** Fornece métricas como precisão, recall e F1-score para cada classe.
    * **Precisão:** Dos produtos classificados como pertencentes a uma classe, quantos realmente pertencem a ela?
    * **Recall:** Dos produtos que realmente pertencem a uma classe, quantos o modelo classificou corretamente?
    * **F1-score:** Média harmônica entre precisão e recall, útil para balancear as duas métricas.

* **Matriz de Confusão:** Visualiza os acertos e erros do modelo para cada classe, mostrando onde ele está confundindo as classes.

### Exemplo de Interpretação:

Imagine que o seu modelo teve uma acurácia de 80%, mas o relatório de classificação mostra que a classe "SA02961" tem um recall muito baixo (e.g., 30%). Isso significa que o modelo está tendo dificuldade em identificar corretamente os produtos dessa classe, o que pode ser um problema.
